```python
# ZJ_Zhang_W2_assn.py

import csv
# be sure to update the path below to reflect your own environment!!
# also be sure that the code is properly indented after you paste it!
with open('C:/Users/lenovo/Desktop/Analytic Programming/week2/cars-sample35.txt') as csvfile:
    

# Task1
# new list
 
# create 7 blank list
    price_list = []
    Maintenance_cost_list = []
    Number_of_doors = []
    Number_of_passengers = []
    Luggage_capacity = []
    Safety_rating = []
    Classification_of_vehicle = []

# read file
    readCSV = csv.reader(csvfile)
    for row in readCSV:
        
# using append() function to put the data into list        
        price_list.append(row[0])
        
        Maintenance_cost_list.append(row[1])
        
        Number_of_doors.append(row[2])
        
        Number_of_passengers.append(row[3])
        
        Luggage_capacity.append(row[4])
        
        Safety_rating.append(row[5])
        
        Classification_of_vehicle.append(row[6])
        
#show result        
    print('This is price_list:')    
    print(price_list)
    
    print('This is Maintenance_cost_list:') 
    print(Maintenance_cost_list)
    
    print('This is Number_of_doors:')
    print(Number_of_doors)
    
    print('This is Number_of_passengers:')
    print(Number_of_passengers)
    
    print('This is Luggage_capacity:')
    print(Luggage_capacity)
    
    print('This is Safety_rating:')
    print(Safety_rating)
    
    print('This is Classification_of_vehicle:')
    print(Classification_of_vehicle)
    
    
    
# Task2 
# find the list index values of each automobile having a price rating of "med"

task2list = [] #new list to put result

for i in range(0,len(price_list)):   
    if price_list[i] == 'med':    
        #returen index value
        task2list.append(i)
        
print('Here are the index value of each automobile having a price rating of med:')
print(task2list)

# result: [6, 16, 20, 23, 26, 29]


# Task3
# find the "number of passengers" value for each auto having a "price" value of "med"

task3list = []

for i in task2list:
    #return value
    task3list.append(Number_of_passengers[i])
    
print('Here are number of passengers" value for each auto having a "price" value of "med":')
print(task3list)

# result: ['more', '2', '2', '2', '4', '2']


# Task4
# find the index value for each automobile having a price value of "high" 
# and a maintenance value that is not "low".

task4list = []

# using 'i' and 'j' to find the data meets condition
for i in range(0,len(price_list)):
    for j in range(0,len(Maintenance_cost_list)):
# ''&'' means AND, meet three conditions simultaneously
        if (price_list[i] == 'high') & (Maintenance_cost_list[j] != 'low') & (i == j):
            task4list.append(i)
            
print('Here are index value of task4 result:')
print(task4list)
            
# result: [0, 1, 2, 27, 31]



# Task5
# find the index value for each auto having 2 doors and a luggage value of "big"

task5list = []
# using 'i' and 'j' to find the data meets condition
for i in range(0,len(Number_of_doors)):
    for j in range(0,len(Luggage_capacity)):
# ''&'' means AND, meet three conditions simultaneously
        if(Number_of_doors[i] == '2') & (Luggage_capacity[j] == 'big') & (i == j):
            task5list.append(i)
            
print('Here are the index value for each auto having 2 doors and a luggage value of "big":')
print(task5list)

# result: [3, 12, 19, 23]



# Task6
# converting the '5more' values to '5'
# calculate the average number of doors across all 35 autos

task6list = []

for i in range(0,len(Number_of_doors)):
# converting the '5more' values to '5
    if(Number_of_doors[i] == '5more'):
        Number_of_doors[i] = '5'
 
print('new Num_of_door list:')
print(Number_of_doors)
count = float(len(Number_of_doors))

# convert each element to integer
Number_of_doors  = [ int(x) for x in Number_of_doors ]
summing = sum(Number_of_doors)

avg = float (summing / count)

print(avg)

# result:
# new Num_of_door list:
# ['2', '4', '4', '2', '3', '2', '4', '3', '2', '4', '2', '4', '2', '3', '5', '2', '2', '5', '3', '2', '4', '2', '5', '2', '4', '3', '3', '3', '5', '5', '3', '5', '5', '4', '3']
# average: 3.3142857142857145
    
```
